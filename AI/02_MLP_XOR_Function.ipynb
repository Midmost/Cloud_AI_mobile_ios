### 딥-러닝 과정 Mulit Layer Perceptron(MLP) 

## Simple Keras 모델 생성/학습 - XOR Function

```python
# 1. Numpy 가져오기
import numpy as np
```

```python
# 2. 입력/출력 데이터 만들기

x = [[0,0],[0,1],[1,0],[1,1]]
y = [[0],[1],[1],[0]] # 다를 때만 1 같으면 0
```

```python
# 3. Keras 패키지 가져오기
from tensorflow import keras
from tensorflow.keras import layers

print(keras.__version__)
```

```python
# 4. MLP 모델 생성
from keras import Sequential
from keras.layers.core import Dense, Activation
model = keras.Sequential()

# model.add(Dense(input_dim=2, units=2))
model.add(Dense(units = 20, input_dim=2, activation = 'relu'))
model.add(Dense(units = 1, activation = 'sigmoid'))


model.summary()
```

```python
# 5. Compile - Optimizer, Loss function 설정


# sgd = keras.optimizers.SGD(learning_rate=0.1)
adam = keras.optimizers.Adam(learning_rate=0.1)
model.compile(loss ='binary_crossentropy', optimizer=adam)


# model.compile(loss='mean_squared_error', 
#               optimizer='adam', 
#               metrics=['accuracy'])
```

```python
# 6. 학습시키기
history = model.fit(x, y, batch_size=1, epochs=200)
```

```python
# 7. 모델 테스트하기
test_data = np.array([[0,1],[1,0],[0,0]])
result = model.predict(test_data)
print(result)
```
